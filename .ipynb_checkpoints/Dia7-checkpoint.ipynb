```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier #Arboles de decisión
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

bc = datasets.load_breast_cancer()

X_train, X_test, y_train, y_test = train_test_split(bc.data, bc.target)

dt = DecisionTreeClassifier(criterion= 'entropy', random_state=0)

dt.fit(X_train, y_train)

y_p = dt.predict(X_test)

score = accuracy_score(y_test,y_p)

print('Score: ´{}'.format(score))
```

```python
# Exportar el arbol de decisión
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz

bc = datasets.load_breast_cancer()

X_train, X_test, y_train, y_test = \
    train_test_split(bc.data, bc.target)


dt = DecisionTreeClassifier(criterion='entropy', 
                           random_state=0)

dt.fit(X_train, y_train)

y_p = dt.predict(X_test)

score = accuracy_score(y_test, y_p)

print('Score: {}'.format(score))

export_graphviz(dt, out_file='tree_bc.dot', class_names=bc.target_names,
                feature_names=bc.feature_names)
```

```python
#comparar arbol de decisión con randomforest
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.tree import export_graphviz
from sklearn.ensemble import RandomForestClassifier


bc = datasets.load_breast_cancer()

X_train, X_test, y_train, y_test = \
    train_test_split(bc.data, bc.target)

rf = RandomForestClassifier(criterion='entropy',n_estimators=10,n_jobs=4)
dt = RandomForestClassifier(criterion='entropy',random_state=0)

dt.fit(X_train, y_train)
rf.fit(X_train, y_train)

y_rf_p = rf.predict(X_test)
y_p = dt.predict(X_test)

score = accuracy_score(y_test, y_p)
score_rf = accuracy_score(y_test, y_rf_p)

print('Score: {}'.format(score))
print('Score: {}'.format(score_rf))
```

#### K vecinos próximos

Es el último clasificador que vamos a ver. Es un concepto diferente a los vistos.

Sobre un modelo ya entrenado, cuando llega una nueva muestra busca los K vecinos más próximos y con esos datos predice y clasifica la nueva muestra.

**IMPORTATE** normalizar los datos antes de importarlos, para que las distancias sean homogeneas en todas las dimensiones, al igual que ocurre con otros clasificadores. Llevamos todos los datos al mismo rango, por ejemplo usando la normalización estandar entre -1 y 1.

Normalización: https://es.wikipedia.org/wiki/Normalizaci%C3%B3n_(estad%C3%ADstica)

La librería **ann aproximated nearest neighbours** permite buscar más rápido el más cercano sin hacer un calculo por fuerza bruta que supone mucho calculo computacional.

No hay entrenamiento, con las muestras busco los  vecinos más cercanos.

Módulo http://scikit-learn.org/stable/modules/generated/sklearn.neighbors.KNeighborsClassifier.html


```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split #partimos los datos
from sklearn.metrics import accuracy_score
from plot_regions import plot_decision_regions
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler #para normalizar

iris = load_iris()

X = iris.data[:, 2:4]
y = iris.target

#Divido entre la base y la validación para comprobar que lo hace bien
X_base, X_validation, y_base, y_validation = train_test_split(X,y) 

#normalizamos al depender de distancias
sc = StandardScaler()
sc.fit(X_base)

X_base_std = sc.transform(X_base)
X_validation_std = sc.transform(X_validation)

kn = KNeighborsClassifier(n_neighbors=5)
kn.fit(X_base_std,y_base)

y_p = kn.predict(X_validation_std)

score = accuracy_score(y_validation, y_p)
print('Accuracy: {}'.format(score))

plot_decision_regions(X_base_std, y_base, kn)
```

```python
#Version del profesor
from sklearn.neighbors import KNeighborsClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from plot_regions import plot_decision_regions

iris = load_iris()

X = iris.data[:,2:4]
y = iris.target

X_base, X_validation, y_base, y_validation = train_test_split(X, y)

sc = StandardScaler()
sc.fit(X_base)

X_base_std = sc.transform(X_base)
X_validation_std = sc.transform(X_validation)
    
kn = KNeighborsClassifier(n_neighbors=15)

kn.fit(X_base_std, y_base)

y_p = kn.predict(X_validation_std)
score = accuracy_score(y_validation, y_p)
print("Score: {}".format(score))

plot_decision_regions(X_base_std, y_base, kn)
plt.scatter(X_validation_std[:,0], X_validation_std[:,1],c=y_validation)
```

#### Aprendizaje Reforzado

El aprendizaje reforzado, un agente observa el resultado del entorno y actua sobre él. El entono también me devuelve una recompensa (reward) calculada automáticamente y sirve para evaluar lo bien o mal que lo hace el agente.

```python
#ejemplo de entrenamiento usando 100 vectores aleatorios y saco el mejor
import numpy as np

env = gym.make('CartPole-v0') #creo el entorno

reward_max = -1

w_max = None

for episode in range (100):
    observation = env.reset() #reseteo el entorno y cojo la observacin ininicial
    w = 2*np.random.rand(4)-1 #vector de pesos aleatorio al inicio
    total_reward = 0
    # vamos a aplicar perceptron
    for _ in range(1000):
        total_reward += 1
        env.render()
        if np.dot(w,observation) > 0: #producto de dos arrays
            action = 0
        else:
            action = 1

        observation, reward, done, info = env.step(action) #doy un paso
        if done:
            if total_reward > reward_max:
                reward_max = total_reward
                w_max = w
            print('Puntuación total: {}'.format(total_reward))
            break
print(w_max,reward_max)
```

```python
import gym
import numpy as np

env = gym.make('CartPole-v0') #creo el entorno

reward_max = -1

w_max = None

total = 0

w = np.matrix([0.2,-0.5,-0.9,-0.9])
#Parto de un valor inicial que se que funciona bien al probar anteriormente

for episode in range (20):

    observation = env.reset() #reseteo el entorno y cojo la observacin ininicial
    total_reward = 0
    # vamos a aplicar perceptron
    for _ in range(1000):
        total_reward += 1
        env.render()
        if np.dot(w,observation) > 0: #producto de dos arrays
            action = 0
        else:
            action = 1
        observation, reward, done, info = env.step(action) #doy un paso
        if done:
            if total_reward > reward_max:
                reward_max = total_reward
                w_max = w
                #actualizo w
                w = w + np.matrix([0.1])
            else:
                w = w - np.matrix([0.1])
            print('Puntuación total: {}'.format(total_reward))
            break
        else:
            reward_max = total_max
            w_max = w
    total += total_reward
print(w_max,reward_max)
print('Media: {}'.format(total/20))
```
