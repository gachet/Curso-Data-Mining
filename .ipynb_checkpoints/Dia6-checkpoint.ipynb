#### Clasificador Logistico

El perceptron tiene problemas de convergencia y con este clasificador lo solucionamos.

Usamos la API y normalizamos los datos de entrada (datos x)



```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
from plot_regions import plot_decision_regions
from matplotlib import pyplot as plt
```

```python
iris = load_iris()

X = iris.data[:, 2:4]
y = iris.target

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, random_state=0)
    
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

lr = LogisticRegression(C=1000, random_state=0)

lr.fit(X_train_std, y_train)

y_p = lr.predict(X_test_std)
score = accuracy_score(y_test, y_p)
print('Accuracy: {}'.format(score))

plot_decision_regions(X_train_std, y_train, lr)

plt.title('Clasificador logístico, score: {}'.format(score))
```

```python
iris = load_iris()

X = iris.data[:, 2:4]
y = iris.target

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, random_state=0)
    
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

lr = LogisticRegression(C=10e-05, random_state=0)

lr.fit(X_train_std, y_train)

y_p = lr.predict(X_test_std)
score = accuracy_score(y_test, y_p)
print('Accuracy: {}'.format(score))

plot_decision_regions(X_train_std, y_train, lr)

plt.title('Clasificador logístico, score: {}'.format(score))
```

```python
iris = load_iris()

X = iris.data[:, 2:4]
y = iris.target

X_train, X_test, y_train, y_test = \
    train_test_split(X, y, random_state=0)
    
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

for i in range (-5,6):
    lr = LogisticRegression(C=10**i, random_state=0)

    lr.fit(X_train_std, y_train)

    y_p = lr.predict(X_test_std)
    score = accuracy_score(y_test, y_p)
    print('Accuracy: {}'.format(score))

    plot_decision_regions(X_train_std, y_train, lr)

    plt.title('Clasificador logístico, score: {}'.format(score))
plt.show()
```

Cuando C es pequeño tenemos planos que se alejan poco del origen y al aumentar el modelo el más complejo

Balance entre me equivoco o tengo un modelo más complejo.

## SVM

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from plot_regions import plot_decision_regions
from matplotlib import pyplot as plt
from sklearn.svm import SVC

iris = load_iris()

X = iris.data[:, 2:4]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

svm = SVC(kernel = 'linear', C=1, random_state=0)
svm.fit(X_train_std, y_train)
    
y_p = svm.predict(X_test_std)
score = accuracy_score(y_test, y_p)
print('Accuracy: {}'.format(score))

plot_decision_regions(X_train_std, y_train, svm)

plt.plot(svm.support_vectors_[:,0], svm.support_vectors_[:,1], 'v') #vectores de soporte

plt.title('SVM, score: {}'.format(score))
plt.show()
```

```python
svm.support_vectors_.shape #vectores de soporte para calcular los hiperplanos
#se muestran como triangulos invertidos
```

```python
svm.support_vectors_
```

```python
import matplotlib.pyplot as plt
import numpy as np

np.random.seed(0)

X_xor = np.random.randn(200, 2)
y_xor = np.logical_xor(X_xor[:, 0] > 0, X_xor[:, 1] > 0) 

y_xor = np.where(y_xor, 1, -1)
plt.scatter(X_xor[y_xor == 1, 0], X_xor[y_xor == 1, 1], c='b', marker='x', label='1')
plt.scatter(X_xor[y_xor == -1, 0], X_xor[y_xor == -1, 1], c='r', marker='s', label='-1')
plt.xlim([-3, 3]);
plt.ylim([-3, 3])
plt.legend(loc='best');
plt.tight_layout();
plt.show()
```

Los kernel son funciones que me dan la distancia entre puntos. Un ejemplo de kerner es el rbf. En SVM puede aplicar el kernel rbf en lugar del lineal

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
from plot_regions import plot_decision_regions
from matplotlib import pyplot as plt
from sklearn.svm import SVC

iris = load_iris()

X = X_xor
y = y_xor

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
    
sc = StandardScaler()
sc.fit(X_train)

X_train_std = sc.transform(X_train)
X_test_std = sc.transform(X_test)

svm = SVC(kernel = 'linear', C=1, random_state=0)
svm.fit(X_train_std, y_train)
    
y_p = svm.predict(X_test_std)
score = accuracy_score(y_test, y_p)
print('Accuracy: {}'.format(score))

plot_decision_regions(X_train_std, y_train, svm)

plt.plot(svm.support_vectors_[:,0], svm.support_vectors_[:,1], 'v') #vectores de soporte

plt.title('SVM, score: {}'.format(score))
plt.show()
```

**No funciona, ponemos el rbf**

```python
svm = SVC(kernel = 'rbf', C=1, random_state=0, gamma = 0.2)
svm.fit(X_train_std, y_train)
    
y_p = svm.predict(X_test_std)
score = accuracy_score(y_test, y_p)
print('Accuracy: {}'.format(score))

plot_decision_regions(X_train_std, y_train, svm)

plt.plot(svm.support_vectors_[:,0], svm.support_vectors_[:,1], 'v') #vectores de soporte

plt.title('SVM, score: {}'.format(score))
plt.show()
```

Hay kernel lineal, rbf, polinomial, etc... que se pueden aplicar

**El rbf es el más popular**

gamma es el equivalente a la desviación típica, es decir, cuanto se va a abrir la gausiana (campana de gaus). Pensar que el kernel mide la distancia o similitud

## Arboles de Decisión

Son clasificadores

El reto es como construir esos árboles

Separar entrenamienton y test

La normalización no nos va a afectar aquí, incluso puede perjudicar a la hora de interpre el arbol

Normalizar es convertir a valores entre -1 y 1 para aplicar los coeficientes


** Con iris**

x1 = longitud del pétalo

x2 = longitud del sépalo

Si longitud del pétalo > 0.5 (hiperplano vertical)

Si longitud del sépalo > 0.6 (hiperplano horizontal)

Las fronteras van a ser combinaciones de planos horizontales o verticales

Como siempre **fit para entrenar** y **predict para evaluar**

#### Instalar graphviz:

http://www.graphviz.org/

pydotplus permite llamar a graphviz desde python: pip install pydotplus

```python
import pydotplus
```

```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from plot_regions import plot_decision_regions
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import export_graphviz
import pydotplus

iris = load_iris()

X = iris.data[:, 2:4]
y = iris.target

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)

dt = DecisionTreeClassifier(criterion='entropy', random_state=0, max_depth=3)

dt.fit(X_train,y_train)

y_p = dt.predict(X_test)
score = accuracy_score(y_p, y_test)
print("Score: {}".format(score))
plot_decision_regions(X,y,dt)
plt.title("DecisionTree score: {}".format(score))
plt.show()

export_graphviz(dt, out_file='iris_tree.dot')

g = pydotplus.Graph(graph_name='iris_tree.dot')

#g.write_png('iris_tree.png')
```
